# Processing X-ray tomography images with Python

[X-ray tomography](https://en.wikipedia.org/wiki/CT_scan) is an imaging technique that produces 3-D images of a scanned object. For most applications of tomography such as medical imaging or materials science, one often wishes to extract and label objects of interest from the 3-D tomography image.

This tutorial is an example of segmentation of 3-D tomography images, using the [``scikit-image``](http://scikit-image.org/) Python package. Most image processing functions of ``scikit-image`` are compatible with 2-D as well as 3-D images, which makes it a tool of choice for processing tomography images. Furthermore, ``scikit-image`` is part of a larger ecosystem of Scientific Python packages, so that it is possible to use other packages, such as Mayavi for 3-D visualization.

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

First, let us load the image data into a NumPy array. The image is stored in raw format, hence we use ``np.fromfile``. It would also be possible to use ``np.memmap`` to open a *memory view*, that is, to create an array object without loading the data into the memory (data are then read from the disk, which is much slower than from the memory, though).

If you wish to execute this notebook, the data file can be downloaded from 
http://sd-27941.dedibox.fr/Al8Cu_1000_g13_t4_200_250.vol

```python
dat = np.fromfile('Al8Cu_1000_g13_t4_200_250.vol', 
                  dtype=np.float32).reshape((50, 1104, 1104))
```

```python
plt.imshow(dat[10], cmap='gray', vmin=0); plt.colorbar()
```

The scanned object is a drilled cylinder of a metallic alloy with two phases (an aluminium-copper alloy, data courtesy of [Pierre Lhuissier](http://www.grenoble-inp.fr/pierre-lhuissier--431155.kjsp) and the [ESRF](http://www.esrf.eu/) synchrotron). We have plotted above a horizontal slice through the 3-D image.

The goal of this tutorial is to label and visualize pixels of the light phase.

Since ``scikit-image`` algorithms require floating-point images to have values between 0 and 1, we first rescale the range of values. Before rescaling, we saturate the darkest and lightest 0.5% of gray values, since these extreme values are due to experimental noise here, and saturating them will result in a better contrast. 

```python
from scipy import stats
vmin, vmax = stats.scoreatpercentile(dat, (0.5, 99.5))
```

```python
dat = np.clip(dat, vmin, vmax)
dat = (dat - vmin) / (vmax - vmin)
plt.imshow(dat[10], cmap='gray')
```

We will perform the segmentation in several steps: first separate the sample from the background (that corresponds to darker pixels aroung the cylinder), and then separate the two phases inside the sample.

As a first step, we apply a denoising filter to the image so that it will be easier to separate the phases using their gray values. Here we use a [bilateral filter](https://en.wikipedia.org/wiki/Bilateral_filter), that reduces noise while preserving edges.

```python
from skimage import restoration
bilateral = restoration.denoise_bilateral(dat)
```

Plotting histograms of gray values shows that peaks of the histogram are better separated in the output of the bilateral filter. However, there is still an overlap between gray values corresponding to the background and gray values of the darkest (aluminium) phase. 

```python
from skimage import exposure
hi_dat = exposure.histogram(dat)
hi_bilateral = exposure.histogram(bilateral)
plt.plot(hi_dat[1], hi_dat[0], label='data')
plt.plot(hi_bilateral[1], hi_bilateral[0],
         label='bilateral')
plt.xlim(0, 0.5)
plt.legend()
plt.title('Histogram of voxel values')
```

Let us create a boolean mask by thresholding the image. Then we apply several [mathematical morphology](https://en.wikipedia.org/wiki/Mathematical_morphology) operations in order to clean the binary images, such as removing holes inside the cylinder, or applying opening and closing of the binary objects to keep only a "compact" cylinder. 

The segmentation method shown here (applying a fast denoising filter, then thresholding and cleaning the mess afterwards) is a bit crude, but since we're mostly interested in segmenting the background in order to get rid of it, we don't care so much about this step. We'll see a more elegant segmentation method below, for the separation of the two phases inside the sample.

```python
sample = bilateral > 0.2
plt.imshow(sample[10], cmap='gray')
```

```python
from scipy import ndimage
sample = ndimage.binary_fill_holes(sample)
plt.imshow(sample[10], cmap='gray')
```

```python
from skimage import morphology
open_object = morphology.opening(sample, morphology.ball(3))
close_object = morphology.closing(open_object, morphology.ball(3))
plt.imshow(close_object[10], cmap='gray')
```

Now that we have determined which pixels belong to the sample, we don't need to keep so much background around it. We keep only the box that encloses tightly the cylinder.

```python
bbox = ndimage.find_objects(close_object)
```

```python
mask = close_object[bbox[0]]
```

```python
roi_dat = dat[bbox[0]]
plt.imshow(roi_dat[10], cmap='gray')
plt.contour(mask[10])
```

## Non-local means and total-variation denoising

In order to separate the two phases, we use a different denoising filter, called non-local means denoising. In addition to preserving edges, non-local means is also good at preserving textures. However, it is much slower than the bilateral filter. Here we apply it only to a subvolume of the sample (if I wanted to perform the segmentation on the whole volume, I would typically loop over subvolumes of the image, and use several processors for the different subvolumes). The non-local means algorithm has several parameters,  please look at the documentation to know how to set them. 

**Warning**: the non-local means algorithm will be available in ``scikit-image`` 0.12 that has not been released yet. See the instructions [here](http://scikit-image.org/docs/stable/install.html#installation-from-source) to install the development version if you wish to use non-local means denoising. Alternatively, you can use the total-variation denoising filter, as shown below.

```python
nlm = restoration.denoise_nl_means(roi_dat[:, 
                                           250:450,
                                           250:450],
                    patch_size=5, patch_distance=7,
                    h=0.12, multichannel=False)
```

The result of the non-local means denoising is very satisfying, with the dark (aluminium) phase that is very smooth, while more subtle gradients in the light (copper) phase that are well preserved. 

```python
plt.imshow(nlm[10], cmap='gray')
```

An alternative to non-local means denoising is too use total-variation denoising, that tends to produce piecewise-constant images.

```python
tv = restoration.denoise_tv_chambolle(roi_dat[:, 250:450, 250:450],
                                      weight=0.2)
plt.imshow(tv[10], cmap='gray')
```

## Segmentation with random walker

Since we are interested in a precise segmentation of the two phases, we will use a segmentation method that uses more information than just the gray values of pixels (as in thresholding), but also spatial information: pixels close to each other are indeed more likely to belong to the same phase.

To do so, we first create an array of markers, corresponding here to pixels for which the gray value is enough to attribute them to one of the phases, that is with gray values on the two sides of the histogram (mid-range values are still undetermined). 

```python
hi_nlm = exposure.histogram(nlm)
hi_tv = exposure.histogram(tv)
```

```python
plt.plot(hi_nlm[1], hi_nlm[0])
plt.plot(hi_tv[1], hi_tv[0])
plt.axvline(0.33, color='k', ls='--')
plt.axvline(0.4, color='k', ls='--')
```

```python
markers = np.zeros(nlm.shape, dtype=np.uint8)
markers[nlm > 0.4] = 1
markers[nlm < 0.33] = 2
```

```python
plt.imshow(markers[10], cmap='gray')
#plt.contour(markers[10], [0.5, 1.5])
```

Now we use the random walker algorithm, that will let the markers diffuse inside the ``nlm`` image, diffusion being more difficult in regions in high gradients (so that it will be difficult for a marker of one the two phases to cross a boundary to diffuse into the other phase). The parameter ``beta`` controls the amount by which gradients are penalized.

```python
from skimage import segmentation
rw = segmentation.random_walker(nlm, markers, beta=1000., mode='cg_mg')
```

```python
plt.imshow(nlm[10], cmap='gray')
plt.contour(rw[10], [1.5])
```

We finish the segmentation process by removing very small objects (200 pixels correspond to objects smaller than 6 pixels), since visual inspection reveals that they do not correspond to true regions of the light phase.

```python
clean_segmentation = morphology.remove_small_objects(rw == 1, 200)
```

Labeling connected component reveals that there is a single connected phase, percolating through the whole image.

```python
from skimage import measure
labels = measure.label(clean_segmentation, 
                        background=0)
print(labels.max())
labels = (labels + 1).astype(np.uint8)
```

## 3-D visualization with mayavi

In order to visualize the 3-D morphology of the segmented phase, we use the [Mayavi](http://docs.enthought.com/mayavi/mayavi/) package. Mayavi is based on VTK: we first have to create a mayavi object from our NumPy array by calling ``mlab.pipeline.scalar_field``. All the other operations written below can also be realized within the graphical pipeline of Mayavi, but we reproduce here the code so that the figure can be reproduced. A specificity of Mayavi is indeed that it is possible to use both the graphical pipeline (choosing filters and visualizations modules from menus), and Python code in a script.

Here we represent the surface of the light copper phase, color-coded by its Gaussian curvature (with round tips having a positive Gaussian-curvature, while saddle-like regions have a negative Gaussian curvature). We first smooth the surface to avoid staircasing effects due to pixels. While this method is well adapted to visualizations, it is not a rigorous computation of local curvatures, however. 

In this notebook, we could not include the 3-D interactive visualization figure of Mayavi (where one can for example rotate the view, zoom, pan, etc.), but running this notebook will create the window (please uncomment the ``mlab.show`` line below before).

```python
from mayavi import mlab
```

```python
mlab.figure(bgcolor=(0, 0, 0))
s = mlab.pipeline.scalar_field(labels)
contour = mlab.pipeline.contour(s)
smooth = mlab.pipeline.user_defined(contour, filter='SmoothPolyDataFilter')
smooth.filter.number_of_iterations = 400
smooth.filter.relaxation_factor = 0.015
curv = mlab.pipeline.user_defined(smooth, filter='Curvatures')

surf = mlab.pipeline.surface(curv)
module_manager = curv.children[0]
module_manager.scalar_lut_manager.data_range = np.array([-0.6,  0.5])
module_manager.scalar_lut_manager.lut_mode = 'RdBu'
#mlab.show()
```
